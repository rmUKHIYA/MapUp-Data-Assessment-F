# Q1. distance matrix

```python


def calculate_distance_matrix(file_path):
    
    df = pd.read_csv(file_path)

    
    distance_matrix = df.pivot_table(index='id_start', columns='id_end', values='distance')

   
    num_rows, num_cols = df.shape
    for i in range(min(num_rows,num_cols)):
        distance_matrix.iloc[i,i]=0

    distance_matrix.fillna(0, inplace=True)

    return distance_matrix

```

```python
file_path = 'dataset-3.csv'
result_matrix = calculate_distance_matrix(file_path)
print(result_matrix)
```

```python


def calculate_distance_matrix(csv_file):
    
    df = pd.read_csv(csv_file)

    
    G = nx.DiGraph()

    
    for _, row in df.iterrows():
        G.add_edge(row['id_start'], row['id_end'], distance=row['distance'])
        G.add_edge(row['id_end'], row['id_start'], distance=row['distance'])  

    
    distance_matrix = nx.floyd_warshall_numpy(G)

    
    distance_df = pd.DataFrame(distance_matrix, index=G.nodes, columns=G.nodes)

    return distance_df

```

```python
csv_file_path = 'dataset-3.csv'
result_df = calculate_distance_matrix(csv_file_path)


print(result_df)
```

# q2.unroll distance matrix

```python

def unroll_distance_matrix(distance_df):
   
    columns = distance_df.columns
    index = distance_df.index

    
    id_start_list = []
    id_end_list = []
    distance_list = []

    for id_start in index:
        for id_end in columns:
            if id_start != id_end:  
                id_start_list.append(id_start)
                id_end_list.append(id_end)
                distance_list.append(distance_df.loc[id_start, id_end])

    
    unrolled_df = pd.DataFrame({
        'id_start': id_start_list,
        'id_end': id_end_list,
        'distance': distance_list
    })

    return unrolled_df


unrolled_df = unroll_distance_matrix(result_df)


print(unrolled_df)
```

```python
unrolled_df = unroll_distance_matrix(result_df)
```

# Q3 Finding IDs within Percentage Threshold

```python


def find_ids_within_ten_percentage_threshold(unrolled_df, reference_value):

    reference_rows = unrolled_df[unrolled_df['id_start'] == reference_value]

    
    average_distance = reference_rows['distance'].mean()


    lower_threshold = average_distance * 0.9
    upper_threshold = average_distance * 1.1

    within_threshold_rows = unrolled_df[
        (unrolled_df['distance'] >= lower_threshold) &
        (unrolled_df['distance'] <= upper_threshold)
    ]

   
    result_ids = sorted(within_threshold_rows['id_start'].unique())

    return result_ids

```

```python
reference_values=unrolled_df["id_start"]
result_ids = find_ids_within_ten_percentage_threshold(unrolled_df,reference_values )


print("id with 10% thersold  : \n\n")
print(result_ids)
```

# Q4.Calculate toll rate 

```python


def calculate_toll_rate(distance_df):
    
    rate_coefficients = {'moto': 0.8, 'car': 1.2, 'rv': 1.5, 'bus': 2.2, 'truck': 3.6}

    
    for vehicle_type, rate_coefficient in rate_coefficients.items():
        distance_df[vehicle_type] = distance_df['distance'] * rate_coefficient

    return distance_df


```

```python
result_df_with_toll_rates = calculate_toll_rate(unrolled_df)

print(result_df_with_toll_rates)
```
